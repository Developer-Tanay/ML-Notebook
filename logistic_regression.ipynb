```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("datasets/social_network_ads.csv")
dataset.drop(columns=["EstimatedSalary"], inplace=True)
dataset.head(3)
```

#### First of all we need to chceck if logistic regression is applicable on this dataset or not

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x="Age", y="Purchased", data=dataset, hue="Purchased")
plt.show()
```

```python
x = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[40]]) # Predicting for a 30-year-old
```

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x="Age", y="Purchased", data=dataset, hue="Purchased")
sns.lineplot(x=x_test["Age"], y=lr.predict(x_test), color="red", label="Logistic Regression")
plt.xlabel("Age")
plt.ylabel("Purchased")
plt.title("Logistic Regression Prediction")
plt.legend()
plt.show()
```
