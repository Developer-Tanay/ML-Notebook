```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("datasets/placement-dataset.csv")
dataset.head(3)
```

```python
# seperate the dependent and independent values
x = dataset.iloc[:, :-1]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100 # accuracy score
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
# we will need training data and testing data for confusion matrix
confusion_matrix = confusion_matrix(y_test, lr.predict(x_test))
confusion_matrix
```

```python
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Not Placed", "Placed"], yticklabels=["Not Placed", "Placed"])
plt.show()
```

## Finding the Precision and recall values

```python
precision_score(y_test, lr.predict(x_test))*100 # precision score
```

```python
recall_score(y_test, lr.predict(x_test))*100 # recall score
```

```python
f1_score(y_test, lr.predict(x_test))*100 # f1 score
```

This is not a good model as the precision, recall and f1 score are very low. We always need to miantain high precision and recall and f1 values.
