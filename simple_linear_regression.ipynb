```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv(r"datasets\placement.csv")
dataset.head(3)
```

```python
# check if our dataset follows linearity
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["cgpa"]]
# x.ndim # check the dimension of the dataset (if it is one dimensional then change it to two dimensional by simply adding a new 3rd bracket)
y = dataset["package"]
```

Now we will split our data into two parts  "training" and "testing" 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# from sklearn.linear_model import LinearRegression - as we already imported it above
# LinearRegression is a class in the module sklearn.linear_model. So now we will need to create an object of this class

lr = LinearRegression() # create an object of LinearRegression class
lr.fit(x_train, y_train) # fit the model with training data
```

```python
# y = mx + c 
lr.coef_ # slope of the line
# lr.intercept_ # y-intercept of the line
```

```python
# we get y = 0.57425647 * x - 1.0270069374542108
```

Now our model is trained and fitted. Now its time to test our model.

```python
lr.score(x_test, y_test)*100 # check the accuracy of the model in percentage
```

```python
lr.predict([[6.89]])
```

```python
y = 0.57425647 * 6.89 - 1.0270069374542108
y 
```

that's how it works

```python
y_prd = lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.plot(dataset["cgpa"], y_prd, color="red")
plt.legend(["org data", "predicted line"])
# plt.savefig("predict.jpg")
plt.show()
```
