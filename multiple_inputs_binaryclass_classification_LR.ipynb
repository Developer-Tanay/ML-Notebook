```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
dataset = pd.read_csv("datasets/placement-dataset.csv")
dataset.head(3) 
```

```python
plt.figure(figsize=(5, 4))
sns.scatterplot(x="cgpa", y="iq", data=dataset, hue="placement") # hue is used to color points by placement status. To check if cgpa and iq is linearly separable or not 
plt.title("cgpa vs iq")
plt.show()
```

### seperate the dataset basis on input output

```python
x = dataset.iloc[:, :-1] # all columns except last column
y = dataset["placement"] # last column
# x.ndim, y.ndim # check the dimensions 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100 # accuracy score
```

```python
lr.predict([[5.2, 110]])
```

```python
lr.coef_ , lr.intercept_ # coefficients and intercept of the model
```

### Visualizing decision boundary

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr) # clf is the classifier e.g. our model name "lr"
plt.xlabel("cgpa")
plt.ylabel("iq")
plt.title("Decision Boundary")
plt.show()
```
